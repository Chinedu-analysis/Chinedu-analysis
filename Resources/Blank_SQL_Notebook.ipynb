```python
%%sql

Select *
From information_schema.columns
WHERE table_name = 'customer'
```

```python
%%sql

Select *
From date
Limit 10
```

```python
%%sql

Select
  quantity,
  netprice,
  quantity * netprice as net_revenue
From sales
Limit 10
```

```python
%%sql

Select
  *,
  quantity * netprice * exchangerate as net_revenue
From sales
Limit 10
```

```python
%%sql

Select
  orderdate,
  quantity * netprice * exchangerate as net_revenue
From sales
Where
  Cast(orderdate as date) >= '2020-01-01'
Limit 10
```

```python
%%sql

Select
  s.orderdate,
  s.quantity * s.netprice * s.exchangerate as net_revenue,
  c.givenname,
  c.surname,
  c.countryfull,
  c.continent,
  p.productkey,
  p productname,
  p.categoryname,
  p.subcategoryname
From sales s
Left join customer c on s.customerkey = c.customerkey
Left join product p on s.productkey = p.productkey
Where
  Cast(orderdate as date) >= '2020-01-01'
Limit 10
```

```python
%%sql

Select
  s.orderdate,
  s.quantity * s.netprice * s.exchangerate as net_revenue,
  c.givenname,
  c.surname,
  c.countryfull,
  c.continent,
  p.productkey,
  p productname,
  p.categoryname,
  p.subcategoryname,
  CASE WHEN s.quantity * s.netprice * s.exchangerate > 1000 Then 'High' Else 'Low' End As net_revenue_category
From sales s
Left join customer c on s.customerkey = c.customerkey
Left join product p on s.productkey = p.productkey
Where
  Cast(orderdate as date) >= '2020-01-01'
Limit 10
```
